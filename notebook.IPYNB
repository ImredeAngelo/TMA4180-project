```python
from src.main import *
import numpy as np
import matplotlib.pyplot as plt

def plot(p, l, title="", size=(12, 12), offset=(0.2, 0, 0)):
    '''Plot structure'''
    ax = plt.figure(figsize=size).add_subplot(projection='3d')

    # Plot points
    ax.scatter([x[0] for x in p], [x[1] for x in p], [x[2] for x in p], c='k')

    # Draw cable lines
    for i in range(len(l)):
        for j in range(i, len(l[0])):
            if l[i][j]:
                xs = [p[i][0], p[j][0]]
                ys = [p[i][1], p[j][1]]
                zs = [p[i][2], p[j][2]]
                ax.plot(xs, ys, zs, 'k--')

        # Also label points
        lp = p[i] + offset
        ax.text(lp[0], lp[1], lp[2], ("p" if i < 4 else "x") + f"({i + 1}) = ({p[i][0]}, {p[i][1]}, {p[i][2]})")

    # Make legend, set axes limits and labels
    ax.legend()
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    plt.title(title)
    plt.show()
```

```python
fixed = np.array([(5, 5, 0), (-5, 5, 0), (-5, -5, 0), (5, -5, 0)])  # Fixed points
initial = np.array([(10, 1, 0), (-1, 1, 0), (-6, -1, 0), (6, -1, 0)]) # Initial conditions

x0, x, l = problem_1(fixed, initial)

plot(x0, l, title="Initial Structure")
plot(x, l, title="Found Solution Structure")
```
